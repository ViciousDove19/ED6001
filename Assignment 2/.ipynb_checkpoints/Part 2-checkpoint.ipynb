```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
from scipy.signal.signaltools import wiener
import skimage
from skimage import feature, io
from skimage.color import rgb2gray
```

# Read Filtered Images

```python
img1 = cv2.imread('img1mean.jpg')
img2 = cv2.imread('img2mean.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

_, (plot1, plot2) = plt.subplots(1, 2)
plot1.title.set_text('Image 1')
plot2.title.set_text('Image 2')
plot1.imshow(img1, cmap='gray')
plot2.imshow(img2, cmap='gray')
plt.show()
```

```python
img = img1
#[140:370,440:770] -- dimensions to crop

#img = img2
#[80:310, 620:1030] -- dimensions to crop

plt.imshow(img, cmap='gray', vmin=0, vmax=255)
```

```python
plt.hist(img.ravel(),256,[0,256])
plt.show()
```

# Histogram Equalization

```python
imghe = cv2.equalizeHist(img)
plt.imshow(imghe, cmap='gray', vmin=0, vmax=255)
```

```python
plt.hist(imghe.ravel(),256,[0,256])
plt.show()
```

# Clahe Equalization

```python
clahe = cv2.createCLAHE(clipLimit = 5)
imgcl = clahe.apply(img)
plt.imshow(imgcl, cmap='gray', vmin=0, vmax=255)
```

```python
plt.hist(imgcl.ravel(),256,[0,256])
plt.show()
```

# Canny Edge Detection

```python
from skimage import feature


image = imgcl
# Compute the Canny filter for two values of sigma
edges1 = feature.canny(image)
edges1 = np.where(edges1 == True, 255, edges1)
edges1 = np.where(edges1 == False, 0, edges1)

plt.imshow(edges1, cmap='gray')
```

```python
plt.imshow(edges1[140:370,440:770], cmap='gray')
```

```python
edges2 = feature.canny(image, sigma=3.5)

edges2 = np.where(edges2 == True, 255, edges2)
edges2 = np.where(edges2 == False, 0, edges2)

plt.imshow(edges2, cmap='gray')
```

```python
plt.imshow(edges2[140:370,440:770], cmap='gray')
```

```python

edges3 = feature.canny(image, sigma=4)

edges3 = np.where(edges3 == True, 255, edges3)
edges3 = np.where(edges3 == False, 0, edges3)

plt.imshow(edges3, cmap='gray')
```

```python
plt.imshow(edges3[140:370,440:770], cmap='gray')
```

```python
cropped1 = edges3[140:370,440:770]

filename = 'cropped1.jpg'
#filename = 'img2mean.jpg'
  
cv2.imwrite(filename, cropped1)
```

```python
sobelxy = cv2.Sobel(src=imgcl, ddepth=-1, dx=1, dy=1, ksize=5) 

sobelxy = np.where(sobelxy >= 200, 255, sobelxy)
sobelxy = np.where(sobelxy <= 200, 0, sobelxy)

plt.imshow(sobelxy, cmap='gray')
```

```python
plt.imshow(sobelxy[140:370,440:770], cmap='gray')
```

```python
laplace = cv2.Laplacian(imgcl, ddepth=-1, ksize=3)

laplace = np.where(laplace >= 50, 255, laplace)
laplace = np.where(laplace <= 50, 0, laplace)

plt.imshow(laplace, cmap='gray')
```

```python
plt.imshow(laplace[140:370,440:770], cmap='gray')
```
