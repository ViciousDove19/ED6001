```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
xray_one_image = Image.open("xray_two.png")
xray_one = np.asarray(xray_one_image)
plt.imshow(xray_one, cmap="gray", vmin=0, vmax=255)
```

```python
histogram = np.zeros(256)
height, width = xray_one.shape

for i in range(height):
    for j in range(width):
        histogram[xray_one[j,i]] += 1

plt.plot(histogram)
```

```python
equalized_histogram = np.zeros(256)
normalization_factor = height*width

for i in range(256):
    for j in range(i+1):
        equalized_histogram[i] += (255*histogram[j])/normalization_factor
        
plt.plot(equalized_histogram)
```

```python
equalized_xray_one = np.zeros((height,width))

for i in range(height):
    for j in range(width):
        equalized_xray_one[j,i] = equalized_histogram[xray_one[j,i]]

plt.imshow(equalized_xray_one, cmap="gray", vmin=0, vmax=255)
```

```python
fig,axs = plt.subplots(1,2)
axs[0].imshow(xray_one,cmap='gray')
axs[1].imshow(equalized_xray_one,cmap='gray')
plt.show()
```
